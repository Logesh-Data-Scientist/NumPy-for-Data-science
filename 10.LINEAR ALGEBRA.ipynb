```python
import numpy as np
```

```python
# Sample Data
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

v = np.array([1, 2])
```

 1. np.dot() → Dot product (VERY IMPORTANT)

```python
print(np.dot(v, v)) 

print(np.dot(A, B))
```

2. np.matmul() → Matrix multiplication

```python
print(np.matmul(A, B))
```

3. np.linalg.inv() → Matrix inverse

```python
inv_A = np.linalg.inv(A)

print(inv_A)
```

4. np.linalg.det() → Determinant

```python
print(np.linalg.det(A))
```

 5. np.linalg.eig() → Eigenvalues & eigenvectors

```python
values, vectors = np.linalg.eig(A)

print("Eigenvalues:", values)
print("Eigenvectors:\n", vectors)
```

 6. np.linalg.svd() → Singular Value Decomposition

```python
U, S, Vt = np.linalg.svd(A)

print("U:\n", U)
print("S:", S)
print("Vt:\n", Vt)
```

7. np.linalg.norm() → Vector / matrix norm

```python
print(np.linalg.norm(v)) 

print(np.linalg.norm(A))
```

8. np.linalg.solve() → Solve linear equations

```python
b = np.array([5, 6])

x = np.linalg.solve(A, b)

print(x)
```

9. np.linalg.rank() → Matrix rank

print(np.linalg.matrix_rank(A))

10. np.cross() → Cross product

```python
a = np.array([1, 0, 0])
b = np.array([0, 1, 0])

print(np.cross(a, b)) 
```

```python
import numpy as np
```

```python
a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])
```

```python
result = np.add(a, b)
print(result)
result = np.subtract(b, a)
print(result) 
result = np.multiply(a, b)
print(result) 
result = np.divide(b, a)
print(result)
result = np.power(a, 2)
print(result)
result = np.mod(b, 3)
print(result) 
result = np.sqrt(a)
print(result) 
result = np.exp(a)
print(result)
result = np.log(a)
print(result)
result = np.log2(a)
print(result)
result = np.log10(a)
print(result)
```

```python
arr = np.array([-1, -5, 3, -10])

result = np.abs(arr)
print(result)
```

```python
arr = np.array([1.9, 2.5, 3.1])

print(np.floor(arr))
print(np.ceil(arr)) 
print(np.round(arr))       
print(np.round(arr, 1))
```

```python
arr = np.array([5, 10, 15, 20, 25])

# Keep values between 10 and 20
clipped = np.clip(arr, 10, 20)

print(clipped)  #
```

SAVING & LOADING DATA

    np.save()           # save array as .npy
    np.load()           # load .npy file
    np.savetxt()        # save as .txt or .csv
    np.loadtxt()        # load from .txt or .csv
    np.genfromtxt()     # load with missing value handling

```python
import numpy as np
```

```python
data = np.array([[1, 2, 3],
                 [4, 5, 6]])
```

1. np.save() → Save as .npy (binary, fast)

```python
np.save("data.npy", data)
```

2. np.load() → Load .npy

```python
loaded = np.load("data.npy")

print(loaded)
```

3. np.savetxt() → Save as text/CSV

```python
np.savetxt("data.txt", data, delimiter=",")
```

4. np.loadtxt() → Load text/CSV

```python
loaded = np.loadtxt("data.txt", delimiter=",")

print(loaded)
```

5. np.genfromtxt() → Load with missing values

```python
loaded = np.genfromtxt("data.txt", delimiter=",", filling_values=0)

print(loaded)
```

    arr.shape           # dimensions of array
    arr.ndim            # number of dimensions

```python
import numpy as np
```

creating a sample array

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print(arr)
```

1. arr.shape → Dimensions of array

```python
print(np.shape(arr))
```

2. arr.ndim → Number of dimensions

```python
print(arr.ndim)
```

3. arr.size → Total number of elements

```python
print(arr.size)
```

4. arr.dtype → Data type of elements

```python
print(arr.dtype)
```

5. arr.itemsize → Memory per element

```python
print(arr.itemsize)
```

6. len(arr) → Length of first dimension

```python
print(len(arr))
```

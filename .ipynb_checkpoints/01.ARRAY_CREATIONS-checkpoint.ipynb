Importing numpy

```python
import numpy as np
```

1. np.array() → Create array from list

```python
arr = np.array([1, 2, 3, 4, 5])

print(arr)       
print(type(arr)) 
```

2. np.zeros() → Array of zeros

```python
zeros_arr = np.zeros((2, 3))

print(zeros_arr)
```

3. np.ones() → Array of ones

```python
ones_arr = np.ones((3, 3))

print(ones_arr)
```

4. np.full() → Array filled with a value

```python
full_arr = np.full((2, 2), 7)

print(full_arr)
```

5. np.arange() → Evenly spaced values (like range)

```python
arr_range = np.arange(10)

arr_step = np.arange(2, 10, 2)

print(arr_range)  
print(arr_step)
```

6. np.linspace() → Fixed number of evenly spaced values

```python
lin_arr = np.linspace(0, 1, 5)

print(lin_arr)
```

7. np.eye() → Identity matrix

```python
identity = np.eye(3)

print(identity)
```

8. np.random.rand() → Random values (0 to 1)

```python
rand_arr = np.random.rand(2, 2)

print(rand_arr)
```

9. np.random.randn() → Normal distribution

```python
randn_arr = np.random.randn(2, 3)

print(randn_arr)
```

10. np.random.randint() → Random integers

```python
randint_arr = np.random.randint(1, 10, size=5)

print(randint_arr)
```

11. np.empty() → Uninitialized array (⚡ very fast)

```python
empty_arr = np.empty((2, 2))

print(empty_arr)
```

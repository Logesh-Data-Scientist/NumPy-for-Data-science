```python
import numpy as np
```

 Sample Data (with NaN & Inf)

```python
arr = np.array([10, 20, np.nan, 40, np.inf, -np.inf])
```

 1. np.isnan() → Detect NaN valuesa

```python
print(np.isnan(arr))
```

 2. np.isinf() → Detect infinite values

```python
print(np.isinf(arr))
```

3. np.isfinite() → Detect valid (finite) values

```python
print(np.isfinite(arr))
```

4. np.nan_to_num() → Replace NaN & Inf

```python
cleaned = np.nan_to_num(arr)

print(cleaned)
```

5. np.nanmean() → Mean ignoring NaN

```python
arr2 = np.array([10, 20, np.nan, 40])

print(np.nanmean(arr2))
```

6. np.nanmedian() → Median ignoring NaN

```python
print(np.nanmedian(arr2)) 
```

7. np.nanstd() → Std ignoring NaN

```python
print(np.nanstd(arr2))
```

8. np.nansum() → Sum ignoring NaN

```python
print(np.nansum(arr2))  
```

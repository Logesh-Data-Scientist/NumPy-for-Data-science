```python
import numpy as np
```

```python
arr = np.array([10, 20, 30, 40, 50])
arr2 = np.array([1, 2, 3, 4, 5])
```

```python
print(np.mean(arr))  
print(np.median(arr))  
print(np.std(arr))
print(np.var(arr))
print(np.min(arr))
print(np.max(arr))
print(np.sum(arr))
print(np.cumsum(arr))
print(np.cumprod(arr2))
print(np.percentile(arr, 50))
print(np.percentile(arr, 25))
print(np.quantile(arr, 0.5)) 
```

```python
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 30, 40])

print(np.corrcoef(x, y))
print(np.cov(x, y))
```

```python
data = np.array([1,2,2,3,3,3,4,4,5])

hist, bins = np.histogram(data, bins=3)

print(hist) 
print(bins) 
```

```python
data = np.array([1, 2, 2, 3, 3, 3])

print(np.bincount(data))
```

```python
data = np.array([1,2,2,3,3,3])

values, counts = np.unique(data, return_counts=True)

print(values)  
print(counts) 
```

```python
values = np.array([10, 20, 30])
weights = np.array([1, 2, 3])

print(np.average(values, weights=weights))
```

ARRAY RESHAPING & MANIPULATION
    
    pythonarr.reshape()       # change shape
    arr.flatten()       # collapse to 1D (copy)
    arr.ravel()         # collapse to 1D (view)
    arr.T               # transpose
    np.transpose()      # transpose with axis control
    np.expand_dims()    # add new dimension
    np.squeeze()        # remove single dimensions
    np.resize()         # resize array

```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])
print(arr)
```

1. arr.reshape() → Change shape

```python
reshaped = arr.reshape(2, 3)

print(reshaped)
```

2. arr.flatten() → Convert to 1D (COPY)


```python
flat = reshaped.flatten()

print(flat)  

flat[0] = 100
print(reshaped) 
```

3. arr.ravel() → Convert to 1D (VIEW)

```python
ravel_arr = reshaped.ravel()

print(ravel_arr)

ravel_arr[0] = 999
print(reshaped)  
```

4. arr.T → Transpose (quick)

```python
matrix = arr.reshape(2, 3)

print(matrix.T)
```

5. np.transpose() → Advanced transpose

```python
print(np.transpose(matrix))

arr3d = np.array([[[1,2],[3,4]]])

transposed = np.transpose(arr3d, (0, 2, 1))
print(transposed)
```

6. np.expand_dims() → Add dimension

```python
expanded = np.expand_dims(arr, axis=0)

print(expanded.shape)  

expanded_col = np.expand_dims(arr, axis=1)
print(expanded_col.shape)  
```

7. np.squeeze() → Remove single dimensions

```python
arr2 = np.array([[[1, 2, 3]]])  # shape (1,1,3)

squeezed = np.squeeze(arr2)

print(squeezed.shape)  # (3,)
```

8. np.resize() → Resize array

```python
resized = np.resize(arr, (3, 4))

print(resized)
```

```python
import numpy as np
```

1. np.random.seed() → Reproducibility (VERY IMPORTANT)

```python
np.random.seed(42)

print(np.random.rand(3))
```

2. np.random.rand() → Uniform (0 to 1)

```python
print(np.random.rand(2, 2))
```

3. np.random.randn() → Standard Normal (mean=0, std=1)

```python
print(np.random.randn(2, 3))
```

4. np.random.randint() → Random integers

```python
print(np.random.randint(1, 10, size=5))
```

 5. np.random.choice() → Random sampling

```python
arr = np.array([10, 20, 30, 40])

print(np.random.choice(arr, size=2))
```

6. np.random.shuffle() → Shuffle in place

```python
arr = np.array([1, 2, 3, 4])

np.random.shuffle(arr)

print(arr)  
```

7. np.random.permutation() → Shuffled copy

```python
arr = np.array([1, 2, 3, 4])

new_arr = np.random.permutation(arr)

print(new_arr)  
print(arr)   
```

8. np.random.normal() → Normal distribution with params

```python
print(np.random.normal(50, 10, 5))
```

9. np.random.uniform() → Uniform with range

```python
print(np.random.uniform(5, 10, 5))
```

10. np.random.binomial() → Binomial distribution

```python
print(np.random.binomial(10, 0.5, 5))
```

11. np.random.poisson() → Poisson distribution

```python
print(np.random.poisson(3, 5))
```

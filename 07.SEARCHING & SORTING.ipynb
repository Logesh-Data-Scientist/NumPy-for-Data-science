```python
import numpy as np
```

```python
arr = np.array([40, 10, 30, 20, 50])
```

1. np.sort() → Sort array

```python
sorted_arr = np.sort(arr)

print(sorted_arr) 
```

2. np.argsort() → Indices that sort array

```python
indices = np.argsort(arr)

print(indices)  
```

3. np.argmax() → Index of max value

```python
print(np.argmax(arr)) 
```

4. np.argmin() → Index of min value

```python
print(np.argmin(arr))  
```

5. np.where() → Conditional selection (VERY IMPORTANT)

```python
result = np.where(arr > 25)

print(result) 
```

```python
new_arr = np.where(arr > 25, 1, 0)

print(new_arr)
```

6. np.searchsorted() → Find insert position

```python
sorted_arr = np.array([10, 20, 30, 40])
```

```python
pos = np.searchsorted(sorted_arr, 25)

print(pos)  
```

7. np.extract() → Extract elements by condition

```python
result = np.extract(arr > 25, arr)

print(result)
```

8. np.nonzero() → Indices of non-zero elements

```python
arr2 = np.array([0, 1, 0, 2, 3])

print(np.nonzero(arr2))
```

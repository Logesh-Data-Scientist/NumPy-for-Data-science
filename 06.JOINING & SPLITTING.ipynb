```python
import numpy as np
```

```python
a = np.array([[1, 2],
              [3, 4]])

b = np.array([[5, 6],
              [7, 8]])
```

1. np.concatenate() → Join arrays (flexible)

```python
result = np.concatenate((a, b), axis=0)

print(result)
```

```python
result = np.concatenate((a, b), axis=1)

print(result)
```

2. np.vstack() → Vertical stack (row-wise)

```python
result = np.vstack((a, b))
print(result)
```

3. np.hstack() → Horizontal stack (column-wise)

```python
result = np.hstack((a, b))

print(result)
```

4. np.stack() → Stack along NEW axis 

```python
result = np.stack((a, b), axis=0)

print(result.shape) 
```

SPLITTING ARRAYS

5. np.split() → Equal splits

```python
arr = np.array([1, 2, 3, 4, 5, 6])

result = np.split(arr, 3)
print(result)
```

6. np.vsplit() → Vertical split

```python
arr2 = np.array([[1,2],
                 [3,4],
                 [5,6],
                 [7,8]])

result = np.vsplit(arr2, 2)
print(result)
```

7. np.hsplit() → Horizontal split

```python
result = np.hsplit(arr2, 2)

print(result)
```

➕ MODIFYING ARRAYS

8. np.append() → Append values

```python
arr = np.array([1, 2, 3])

new_arr = np.append(arr, 4)
print(new_arr)
```

```python
new_arr = np.append(arr, [5, 6])
print(new_arr) 
```

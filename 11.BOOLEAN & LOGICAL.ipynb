```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4, 5])
```

1. np.any() → True if ANY element is True

```python
print(np.any(arr > 3)) 
```

2. np.all() → True if ALL elements are True

```python
print(np.all(arr > 0))  # True
print(np.all(arr > 3))  # False
```

3. np.logical_and() → Element-wise AND

```python
result = np.logical_and(arr > 2, arr < 5)

print(result)
```

4. np.logical_or() → Element-wise OR

```python
result = np.logical_or(arr < 2, arr > 4)

print(result)
```

5. np.logical_not() → Element-wise NOT

```python
mask = arr > 3

print(np.logical_not(mask))
```

6. np.equal() → Element-wise ==

```python
print(np.equal(arr, 3))
```

7. np.greater() → Element-wise >

```python
print(np.greater(arr, 3))
```

8. np.less() → Element-wise <

```python
print(np.less(arr, 3))
```
